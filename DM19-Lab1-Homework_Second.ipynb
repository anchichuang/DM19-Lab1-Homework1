### 5.4 Dimensionality Reduction

    Input: Raw term-vector matrix

    Output: Projections

```python
from sklearn.decomposition import PCA
```

```python
i_reduced = PCA(n_components = 2).fit_transform(sentiment_df_imdb_counts.toarray())
y_reduced = PCA(n_components = 2).fit_transform(sentiment_df_yelp_counts.toarray())
a_reduced = PCA(n_components = 2).fit_transform(sentiment_df_amazon_counts.toarray())
```

```python
i_reduced.shape
```

```python
y_reduced.shape
```

```python
a_reduced.shape
```

```python
col = ['coral', 'blue']
categories = ['0','1']
#print(type(sentiment_df_yelp['score']))
```

### Plot PCA value

#### PCA Plot of yelp dataset

```python
import matplotlib.pyplot as plt
%matplotlib inline



fig = plt.figure(figsize = (25,10))
ax = fig.subplots()


for c, category in zip(col, categories):
    xs = y_reduced[sentiment_df_yelp['sentiment'] == category].T[0]
    ys = y_reduced[sentiment_df_yelp['sentiment'] == category].T[1]   
    
    ax.scatter(xs, ys, c = c, marker='o')
    
ax.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
ax.set_xlabel('\nX Label')
ax.set_ylabel('\nY Label')
plt.show()
```
